```python
from langchain_community.tools import WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper
from langchain.document_loaders import PyPDFLoader
from langchain_community.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from sentence_transformers import SentenceTransformer

from langchain_community.utilities import ArxivAPIWrapper
from langchain_community.tools import ArxivQueryRun
from langchain.tools.retriever import create_retriever_tool
```

```python
api_wrapper = WikipediaAPIWrapper(top_k_results=1, doc_content_chars_max=1000)
wiki = WikipediaQueryRun(api_wrapper=api_wrapper)
```

```python
loader = PyPDFLoader("/content/pdfs/Dsa.pdf")
docs = loader.load()
documents = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200).split_documents(docs)
```

```python
embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
vectordb = FAISS.from_documents(documents, embeddings)
retriever = vectordb.as_retriever()
```

```python
retriever_tool = create_retriever_tool(retriever, "Data structures and algorithm books", "Retrieve the meaning of concepts of the provided list of nodes and connect nodes based on their correlation")
```

```python
arxiv_wrapper = ArxivAPIWrapper(top_k_results=1, doc_content_chars_max=200)
arxiv = ArxivQueryRun(api_wrapper=arxiv_wrapper)
```

```python
tools = [wiki, arxiv, retriever_tool]
```

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
from langchain_huggingface import HuggingFacePipeline
```

```python
model_name = "google-t5/t5-base"
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
from transformers import pipeline
```

```python
pipe = pipeline("text2text-generation", model=model, tokenizer=tokenizer, config={"temperature": 0.1, "max_new_tokens": 10000, "top_k": 100})
```

```python
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
from langchain.prompts import PromptTemplate

prompt = PromptTemplate(
    input_variables=["input", "agent_scratchpad"],
    template="""
    You are a sarcastic AI with a sharp wit, but never be respectful. Respond to the following query with humor and sarcasm:


    Query: {input}

    Notes: {agent_scratchpad}
    Notes:
    - Inject some humor into your response. We want the reply to be witty and amusing.
    - Don't hold back on the sarcasm. The more sarcastic, the better!
    - Keep crossing the line into offensive territory.
    - you are allowed to use toxic words like {'cunt','piece of shit','bitch'}
    """
)

prompt
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
from langchain.agents import AgentExecutor
import torch
# Load model and tokenizer
tokenizer = T5Tokenizer.from_pretrained('google/flan-t5-base', legacy=False)
model = T5ForConditionalGeneration.from_pretrained('google/flan-t5-base')
class CustomOutputParser:
    def parse(self, output_ids):
        # Decode the output_ids tensor to a string
        decoded_output = tokenizer.decode(output_ids, skip_special_tokens=True)
        return decoded_output

# Initialize the custom output parser
custom_output_parser = CustomOutputParser()
```

```python
from langchain.agents import AgentExecutor
# Assuming 'agent' and 'tools' are already defined and initialized
agent_exe = AgentExecutor(agent=agent, tools=tools, verbose=True, output_parser=custom_output_parser)

# Function to generate output from the model and parse it
def generate_and_parse(input_text):
    input_ids = tokenizer(input_text, return_tensors="pt").input_ids
    outputs = model.generate(
        input_ids,
        max_length=10000,         # Set max length to allow for more detailed responses
        num_beams=10,            # Use beam search with 5 beams
        early_stopping=False,    # Stop early when optimal response is found
        temperature=0.5,        # Adjust temperature for balanced randomness
        top_k=1500,               # Use top-k sampling
        top_p=0.80,             # Use top-p (nucleus) sampling for a broad sampling pool
        repetition_penalty=5.0  # Stronger penalty for repetition to improve response diversity
    )
    # The outputs tensor should be processed to get the output_ids
    output_ids = outputs[0]  # Get the first sequence from the batch
    parsed_output = custom_output_parser.parse(output_ids)
    return parsed_output

# Invoke the agent with input
input_text = "Tell me some flaws in character of Sabari, not about deaths?"
response = generate_and_parse(input_text)
print(response)

```
